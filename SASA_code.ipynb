```python
from Bio.PDB import PDBParser
from Bio.PDB.SASA import ShrakeRupley

# Create a PDBParser instance
p = PDBParser(QUIET=1)

# Parse the PDB file and store the structure
struct = p.get_structure("P03427", "P03427.pdb")

# Create a ShrakeRupley instance
sr = ShrakeRupley()

# Compute SASA for the entire structure
sr.compute(struct, level="S")

# Iterate through all atoms and print their SASA values
for model in struct:
    for chain in model:
        for residue in chain:
            for atom in residue:
                atom_name = atom.get_id()
                sasa_value = round(atom.sasa, 2)
                print(f"Atom: {atom_name}, SASA: {sasa_value}")
```
