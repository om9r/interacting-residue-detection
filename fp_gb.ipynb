```python
#importing the relevent libraries
import os
import pandas as pd
import numpy as np
import tensorflow as tfa
import tensorflow_decision_forests as tfdf
import seaborn as sns
```

```python
#importing OS environment
#allows you to use the older version of Keras when working with TensorFlow, primarily for compatibility reasons with code written for the standalone Keras library
os.environ['TF_USE_LEGACY_KERAS'] = '1'
```

```python
df = pd.read_excel("spotone_ml.xlsx")
df
```

```python
df['Label'] = df['Label'].replace(np.nan, 0)
df
```

```python
def split_dataset(df, test_ratio=0.20):
  """Splits a panda dataframe in two."""
  test_indices = np.random.rand(len(df)) < test_ratio
  return df[~test_indices], df[test_indices]

train_ds_pd, test_ds_pd = split_dataset(df)
print("{} examples in training, {} examples for testing.".format(
    len(train_ds_pd), len(test_ds_pd)))
```

```python
train_ds = tfdf.keras.pd_dataframe_to_tf_dataset(train_ds_pd, label="Label")
test_ds = tfdf.keras.pd_dataframe_to_tf_dataset(test_ds_pd, label="Label")
```

```python
model_2 = tfdf.keras.GradientBoostedTreesModel()
model_2.fit(train_ds)

print(model_2.summary())
```

```python
#model evaluation
model_2.compile(metrics=["accuracy"])
evaluation = model_2.evaluate(test_ds, return_dict=True)
print()

for name, value in evaluation.items():
  print(f"{name}: {value:.4f}")

#We observe that the accuracy for Gradient Boosted Trees model is higher than that of Random Forest model
```

```python
df1 = pd.read_excel("viral-ml.xlsx")
df1
dftf = tfdf.keras.pd_dataframe_to_tf_dataset(df1)
```

```python
# Assuming 'new_data' is your new dataset
predictions = model_2.predict(dftf)

# If you want to print the predicted labels
print("Predicted Labels:")
print(predictions)
```

```python
csv_file_path = "preds.csv"

# Create a DataFrame with the predicted labels
result_df = pd.DataFrame({'Predicted_Labels': predictions.flatten()})

# Save the DataFrame to a CSV file
result_df.to_csv(csv_file_path, index=False)

# Print a message indicating the successful save
print(f"Predictions saved to {csv_file_path}")
```
